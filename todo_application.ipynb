```python
from tkinter import *
from tkinter import messagebox
import tkinter 
```

```python
#creating the initial window
window=Tk()
#giving a title
window.title("Your To-Do List APP")


#Frame widget to hold the listbox and the scrollbar
frame_task=Frame(window)
frame_task.pack()

#to hold items in a listbox
listbox_task=Listbox(frame_task,bg="black",fg="white",height=15,width=50,font = "Helvetica",selectmode="multiple")  
listbox_task.pack(side=tkinter.LEFT,expand = YES, fill ="both")


#Scrolldown in case the total list exceeds the size of the given window 
scrollbar_task=Scrollbar(frame_task)
scrollbar_task.pack(side=tkinter.RIGHT,fill=tkinter.Y)
listbox_task.config(yscrollcommand=scrollbar_task.set)
scrollbar_task.config(command=listbox_task.yview)

def entertask():
    
    input_task=" "
    def add():
        input_task=entry_task.get(1.0,"end-1c")
        if input_task==" ":
            tkinter.messagebox.showwarning(title="warning", message="please enter a task")
        else:
            listbox_task.insert(END,input_task)
            root1.destroy()
    root1=Tk()
    root1.title("Add Task")
    entry_task=Text(root1, width=40, height=5)
    entry_task.pack()
    button_task=Button(root1,text="Add task",command=add)
    button_task.pack()
    root1.mainloop()
    
def deletetask():
    select = listbox_task.curselection()
    for index in select[::-1]:
        listbox_task.delete(index)

    
    
def markcompleted():
    marked = listbox_task.curselection()
    for index in marked:
        temp = listbox_task.get(index)
        temp_marked = temp + '  (complete)  '
        listbox_task.delete(index)
        listbox_task.insert(index, temp_marked)
    

    


#Button widget 
entry_button=Button(window,text="Add task",width=50,command=entertask,bg="grey",fg="white")
entry_button.pack(pady=3,padx=3)

delete_button=Button(window,text="Delete selected task",width=50,command=deletetask,bg="grey",fg="white")
delete_button.pack(pady=3,padx=3)


mark_button=Button(window,text="Mark as completed ",width=50,command=markcompleted,bg="grey",fg="white")
mark_button.pack(pady=3,padx=3)


window.mainloop()
```
